```python
import pandas as pd

news = pd.read_excel("COIN_sent.xlsx", header=3, index_col=0)
news
```

```python
type(news.index)
```

```python
full_date_range = pd.date_range(start=news.index.min(), end=news.index.max())
news = news.reindex(full_date_range)
news
```

```python
news.ffill(inplace=True)
```

```python
news = news.shift(1)
news.dropna(inplace=True)
```

```python
news.index.name = "date_dt"
news
```

```python
news.to_csv("../Data/BB_sentiment_lag1.csv")
```

```python
news.columns = news.columns.str.replace("_", " ")
news.columns = news.columns.str.replace("TWITTER ", "")
```

```python
import matplotlib.pyplot as plt

axes = news.hist(layout=(3, 3), figsize=(15, 15))
for ax in axes.flatten():
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")
plt.show()
```
